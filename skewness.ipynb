Left Skewness Right Skewness and normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Normal Distribution (mean-centered)
normal_data = np.random.normal(loc=0, scale=1, size=5000)

# Right-Skewed Distribution (positively skewed) – exponential
right_skewed = np.random.exponential(scale=1, size=5000)

# Left-Skewed Distribution – by reversing right skew
left_skewed = -np.random.exponential(scale=1, size=5000)

# Plotting all three
plt.figure(figsize=(10, 6))

plt.hist(normal_data, bins=40, alpha=0.6, label="Normal Distribution")
plt.hist(right_skewed, bins=40, alpha=0.6, label="Right-Skewed")
plt.hist(left_skewed, bins=40, alpha=0.6, label="Left-Skewed")

plt.title("Normal vs Right-Skewed vs Left-Skewed Distributions")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Creating datasets
np.random.seed(0)

# Positive correlation
x1 = np.random.rand(100)
y1 = 2 * x1 + np.random.rand(100) * 0.2

# Negative correlation
x2 = np.random.rand(100)
y2 = -2 * x2 + 1 + np.random.rand(100) * 0.2

# No correlation
x3 = np.random.rand(100)
y3 = np.random.rand(100)

# Plotting
plt.figure(figsize=(12, 4))

# Positive correlation
plt.subplot(1, 3, 1)
plt.scatter(x1, y1, color='green')
plt.title("Positive Correlation")
plt.xlabel("X")
plt.ylabel("Y")

# Negative correlation
plt.subplot(1, 3, 2)
plt.scatter(x2, y2, color='red')
plt.title("Negative Correlation")
plt.xlabel("X")
plt.ylabel("Y")

# No correlation
plt.subplot(1, 3, 3)
plt.scatter(x3, y3, color='blue')
plt.title("No Correlation")
plt.xlabel("X")
plt.ylabel("Y")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -------------------------------
#  FIGURE 1 : SKEWNESS PLOTS
# -------------------------------

# Normal Distribution
normal_data = np.random.normal(loc=0, scale=1, size=5000)

# Right-Skewed (Positive skew)
right_skewed = np.random.exponential(scale=1, size=5000)

# Left-Skewed (Negative skew)
left_skewed = -np.random.exponential(scale=1, size=5000)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.hist(left_skewed, bins=40, color='orange')
plt.title("Left Skewed")

plt.subplot(1, 3, 2)
plt.hist(normal_data, bins=40, color='green')
plt.title("Normal Distribution")

plt.subplot(1, 3, 3)
plt.hist(right_skewed, bins=40, color='blue')
plt.title("Right Skewed")

plt.suptitle("Figure 1: Skewness in Distributions")
plt.tight_layout()
plt.show()


# -------------------------------
#  FIGURE 2 : CORRELATION PLOTS
# -------------------------------

np.random.seed(0)

# Positive correlation
x1 = np.random.rand(100)
y1 = 2 * x1 + np.random.rand(100) * 0.2

# Negative correlation
x2 = np.random.rand(100)
y2 = -2 * x2 + 1 + np.random.rand(100) * 0.2

# No correlation
x3 = np.random.rand(100)
y3 = np.random.rand(100)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(x1, y1, color='green')
plt.title("Positive Correlation")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1, 3, 2)
plt.scatter(x2, y2, color='red')
plt.title("Negative Correlation")
plt.xlabel("X")
plt.ylabel("Y")

plt.subplot(1, 3, 3)
plt.scatter(x3, y3, color='blue')
plt.title("No Correlation")
plt.xlabel("X")
plt.ylabel("Y")

plt.suptitle("Figure 2: Correlation Between Variables")
plt.tight_layout()
plt.show()
```
